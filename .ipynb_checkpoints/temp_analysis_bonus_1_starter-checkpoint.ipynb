### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
# june_df = temp_data_df[temp_data_df['date']=="%06"]
# june_df
temp_data_df['month'] = pd.DatetimeIndex(temp_data_df['date']).month
temp_data_df
```

```python
june_df = temp_data_df[temp_data_df["month"] == 6]
june_df
```

```python
# Identify the average temperature for June
june_df['tobs'].mean()
```

```python
# Identify the average temperature for December
dec_df = temp_data_df[temp_data_df["month"] == 12]
dec_df
dec_df['tobs'].mean()
```

```python
# Create collections of temperature data
june_tobs = temp_data_df[temp_data_df['month'] == 6]['tobs']
dec_tobs = temp_data_df[temp_data_df['month'] == 12]['tobs']
stats.ttest_ind(june_tobs, dec_tobs) 
```

```python
# Run paired t-test
combined_date_df= pd.merge(june_df, dec_df, how = "outer", on=["station", "station"])
combined_date_df

```

```python
stats.ttest_rel(combined_date_df['tobs_x'], combined_date_df['tobs_y'])
```

### Analysis

What we learned here are findings are statiscally significant. The paired t-test makes more sense to use here because they are coming from the same station.
